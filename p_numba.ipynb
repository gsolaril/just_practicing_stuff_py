## Exercise 1 – First @jit Decoration

**Goal**: Compile a simple function and verify the speedup.

**Requirements**:

- Write `sum_squares(n)` in plain Python and measure its time with `%timeit`.
- Decorate it with `@numba.jit(nopython=True)` and measure again after a warm-up call.
- Print the speedup factor.

**Hint**: The first call triggers compilation; always warm up before benchmarking.

## Exercise 2 – Type Annotations and nopython Mode

**Goal**: Use explicit type signatures and understand nopython restrictions.

**Requirements**:

- Decorate `dot_product(a, b)` with `@numba.jit(float64(float64[:], float64[:]))` for eager compilation.
- Try passing a Python list (not ndarray) and observe the error.
- Rewrite with `@numba.jit(nopython=True, cache=True)` and explain what `cache=True` does.

**Hint**: In `nopython` mode, Numba cannot fall back to the interpreter; it forces full compilation.

## Exercise 3 – Sorting Algorithms

**Goal**: Implement and compare JIT-compiled sorting methods.

**Requirements**:

- Write `insertion_sort(arr)` with `@numba.jit(nopython=True)` (in-place, returns None).
- Write `merge_sort(arr)` JIT-compiled.
- Compare their times against `np.sort` for arrays of size 10_000.

**Hint**: Return a new array or sort in-place consistently; avoid Python objects inside jitted functions.

## Exercise 4 – Trading Indicators

**Goal**: Accelerate rolling-window indicator calculations.

**Requirements**:

- Write `rolling_mean(prices, window)` with `@numba.jit(nopython=True)`.
- Write `ema(prices, alpha)` (exponential moving average) JIT-compiled.
- Compare speed against a pandas `.rolling().mean()` and `.ewm().mean()` on 1_000_000 rows.

**Hint**: Pre-allocate the output array with `np.empty_like` before the loop.

## Exercise 5 – Linear System Solver (Jacobi Iteration)

**Goal**: Solve `Ax = b` iteratively with Numba.

**Requirements**:

- Implement one Jacobi iteration step as a JIT-compiled function.
- Run the iteration until convergence (`||x_new - x_old|| < tol`).
- Benchmark against `np.linalg.solve` for a 200x200 system.

**Hint**: Jacobi converges only for diagonally dominant matrices; generate test matrices accordingly.

## Exercise 6 – ODE Solver (Euler and Runge-Kutta)

**Goal**: Integrate a differential equation numerically with JIT speed.

**Requirements**:

- Implement `euler_step(f, t, y, h)` and `rk4_step(f, t, y, h)` as JIT functions.
- Solve `dy/dt = -y` (exponential decay) from t=0 to t=5 and compare to the exact solution `exp(-t)`.
- Plot the error of each method as a function of step size `h`.

**Hint**: Pass the ODE function `f` as a `@jit`-compiled function too, so it can be called from nopython mode.

## Exercise 7 – Best Practices and When to Use Numba

**Goal**: Apply quality guidelines and compare Numba with alternatives.

**Requirements**:

- Profile a numerical loop with `%timeit` (pure Python, then Numba, then numpy).
- List three cases where Numba excels and three where it does not help (or hinders).
- Demonstrate `@numba.vectorize` for element-wise operations and compare with `np.vectorize`.

**Hint**: Numba shines on loops that numpy cannot vectorize natively; it struggles with object dtypes.
