## Exercise 1 – Packaging Terminology

**Goal**: Distinguish key packaging concepts.

**Requirements**:

- Print a table mapping each term to a one-line definition: *import package*, *distribution*, *wheel*, *sdist*, *index*, *pip*.
- Use only the stdlib (no extra imports needed).

**Hint**: Think: what does a user type vs what gets installed vs what Python imports.

## Exercise 2 – Minimal pyproject.toml

**Goal**: Write a valid `pyproject.toml` for a tiny package.

**Requirements**:

- Create a temp directory with `src/mylib/__init__.py`.
- Write a `pyproject.toml` with the `[build-system]` and `[project]` tables filled in.
- Parse it back with `tomllib` (stdlib, Python >= 3.11) and print `name` and `version`.

**Hint**: `tomllib.loads(text)` reads TOML from a string; `importlib.metadata` reads installed packages.

## Exercise 3 – Metadata, Dependencies, and Entry Points

**Goal**: Enrich the project metadata and add a console-script entry point.

**Requirements**:

- Add `dependencies = ["requests>=2.28"]` and an `[extras]` group `dev = ["pytest"]`.
- Add a `[project.scripts]` entry that points to a `mylib.cli:main` function.
- Print the parsed scripts table.

**Hint**: Entry points follow the format `name = "module:callable"`.

## Exercise 4 – Building Wheels and sdists

**Goal**: Build both artifact types locally.

**Requirements**:

- In the project directory from Exercise 2, run `python -m pip wheel --no-deps . -w dist`.
- Inspect the generated `.whl` filename and explain what each segment means.
- If `build` is installed, also run `python -m build --sdist`.

**Hint**: Wheel filename format: `{name}-{version}-{python}-{abi}-{platform}.whl`.

## Exercise 5 – Local Installs and Editable Mode

**Goal**: Install a package locally and verify it works.

**Requirements**:

- In a fresh `venv`, run `pip install -e <path>` for the project from Exercise 2.
- Import `mylib` and call `mylib.__version__` to confirm the install.
- Show the difference between normal install and editable install using `pip show`.

**Hint**: `pip show <name>` displays the install location; editable installs point back to the source tree.

## Exercise 6 – Publishing to TestPyPI

**Goal**: Walk through the full publish workflow without actually uploading.

**Requirements**:

- Print the two-step workflow: `python -m build` then `twine upload --repository testpypi dist/*`.
- Show what `twine check dist/*` does and why it matters.
- Explain the TestPyPI install command including `--extra-index-url`.

**Hint**: `twine check` validates that the README renders correctly on PyPI before you upload.

## Exercise 7 – Authentication and .pypirc

**Goal**: Configure credentials safely.

**Requirements**:

- Write an example `.pypirc` to a temp file with `username = __token__` and a placeholder password.
- Parse it with `configparser` and mask the token (show only first 10 chars).
- List the three recommended auth methods in order of security preference.

**Hint**: Prefer environment variable `TWINE_PASSWORD` over storing tokens in `.pypirc`.

## Exercise 8 – Versioning and Bumping Releases

**Goal**: Automate a patch-version bump.

**Requirements**:

- Write `bump_version(current: str, part: str) -> str` supporting `'major'`, `'minor'`, `'patch'`.
- Apply it to bump `0.3.1` to `0.3.2` (patch), `0.4.0` (minor), `1.0.0` (major).
- Update a `pyproject.toml` string in-memory with the new version using `re.sub`.

**Hint**: SemVer: bumping `minor` resets `patch` to 0; bumping `major` resets both.

## Exercise 9 – Private Indexes and Constraints

**Goal**: Install from alternative sources and pin transitive versions.

**Requirements**:

- Print the `pip install` command that sets a primary index to a hypothetical company URL.
- Write a `constraints.txt` that pins `urllib3<3` and excludes `requests==2.32.0`.
- Explain the dependency-confusion risk of `--extra-index-url` vs `--index-url`.

**Hint**: `--index-url` replaces PyPI; `--extra-index-url` adds an extra index (risk: name shadowing).

## Exercise 10 – Release Management: Yank, Transfer, Remove

**Goal**: Handle post-publish operations.

**Requirements**:

- For each scenario state the correct action and explain why: (a) a broken wheel was uploaded, (b) a token was leaked in a release, (c) ownership should move to a new team, (d) you want to reuse the same version number.
- Write a short helper that prints a decision table for these scenarios.

**Hint**: Yanking hides a release from `pip install` without a version specifier but still allows explicit pinning.
