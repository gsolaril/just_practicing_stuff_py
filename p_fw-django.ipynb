## Exercise 1 тАУ Django Setup and Minimal Configuration

**Goal**: Configure Django programmatically for notebook use.

**Requirements**:

- Call `django.setup()` after setting `DJANGO_SETTINGS_MODULE` or using `settings.configure(...)`.
- Verify the setup with `django.get_version()` and `django.conf.settings.INSTALLED_APPS`.

**Hint**: `settings.configure(...)` lets you bypass the settings file entirely for quick experiments.

## Exercise 2 тАУ Creating a Project and App Structure

**Goal**: Understand the project/app layout.

**Requirements**:

- Print the directory tree produced by `django-admin startproject trading .`.
- Explain what `manage.py`, `settings.py`, `urls.py`, and `wsgi.py` do.
- List the commands to create a `market` app and register it in `INSTALLED_APPS`.

**Hint**: An *app* is a reusable component; a *project* ties multiple apps together.

## Exercise 3 тАУ Models and the ORM

**Goal**: Define models and perform CRUD operations.

**Requirements**:

- Define `Symbol(ticker, name)` and `Order(symbol FK, side, qty, price, created_at)` models.
- Run migrations in-memory with `call_command('migrate')`.
- Create, read (filter), update, and delete `Order` objects using the ORM.

**Hint**: Use `TestCase` from `django.test` for automatic per-test database isolation.

## Exercise 4 тАУ Admin Site Registration

**Goal**: Register models and customize the admin list view.

**Requirements**:

- Register `Symbol` and `Order` with `admin.site.register`.
- Create a custom `OrderAdmin` with `list_display`, `list_filter`, and `search_fields`.
- Use Django's test client to GET `/admin/` and assert the response is 200.

**Hint**: `User.objects.create_superuser(...)` creates a login-ready admin user in tests.

## Exercise 5 тАУ Views and URLs

**Goal**: Map URLs to view functions that return JSON.

**Requirements**:

- Write a `prices(request)` view returning `JsonResponse({'AAPL': 182.5})`.
- Wire it to `/api/prices/` in `urlpatterns`.
- Use `RequestFactory` to call the view without a running server and assert status 200.

**Hint**: `django.test.RequestFactory` creates request objects without the middleware stack.

## Exercise 6 тАУ Class-Based Views

**Goal**: Replace function views with CBVs.

**Requirements**:

- Rewrite `prices` as a `View` subclass overriding `get`.
- Write an `OrderListView(ListView)` that returns all orders as JSON.
- Demonstrate how `as_view()` wires a CBV into `urlpatterns`.

**Hint**: `ListView.get_queryset()` is the standard override point for filtering.

## Exercise 7 тАУ Forms and Validation

**Goal**: Validate user input with a Django Form.

**Requirements**:

- Define `OrderForm(forms.Form)` with `symbol` (CharField), `qty` (IntegerField > 0), and `price` (DecimalField > 0) fields.
- Instantiate with valid and invalid data and call `is_valid()`.
- Print `form.errors` for invalid submissions.

**Hint**: `form.cleaned_data` is only populated after a successful `is_valid()` call.

## Exercise 8 тАУ REST-Style APIs

**Goal**: Build a simple JSON API without DRF.

**Requirements**:

- Write POST `/api/orders/` that reads the JSON body, validates with `OrderForm`, saves to DB, and returns the created order as JSON.
- Write GET `/api/orders/<pk>/` that returns a single order or 404.
- Test both endpoints with Django's `Client`.

**Hint**: `json.loads(request.body)` reads the raw JSON body; wrap in try/except for malformed input.

## Exercise 9 тАУ Settings and Environment Configuration

**Goal**: Separate settings for development and production.

**Requirements**:

- Print a `settings/` structure with `base.py`, `dev.py`, and `prod.py`.
- Show how `dev.py` imports from `base.py` and overrides `DEBUG` and `DATABASES`.
- Demonstrate reading a secret from `os.environ.get('SECRET_KEY', 'fallback')`.

**Hint**: Never hard-code secrets; use environment variables or `django-environ`.

## Exercise 10 тАУ Testing Django Apps

**Goal**: Write a test suite using `django.test.TestCase`.

**Requirements**:

- Write `TestOrderAPI` using `self.client` to POST and GET orders.
- Assert correct HTTP status codes, JSON fields, and database state.
- Use `self.assertContains` and `self.assertJSONEqual` for response assertions.

**Hint**: `django.test.TestCase` wraps each test in a transaction that rolls back automatically.
